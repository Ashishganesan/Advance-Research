```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Convert date and set index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
```

```python
# Plot time series
plt.figure(figsize=(10, 5), dpi=300)
plt.plot(df['Price'], color='brown', label='Oil Price')
# Titles and labels with font size 16
plt.title('Oil Prices (Jul 2020 - Dec 2022)', fontsize=16)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Price ($)', fontsize=16)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Grid
plt.grid(True)

# Add legend in the top-left (upper left inside the plot)
plt.legend(loc='upper left', fontsize=16)

# Annotate the maximum price point
max_price_date = df['Price'].idxmax()
max_price_value = df['Price'].max()

plt.annotate(
    f'Max: {max_price_value:.2f}',
    xy=(max_price_date, max_price_value),
    xytext=(-60, -40),  # offset
    textcoords='offset points',
    arrowprops=dict(facecolor='red', arrowstyle='->'),
    fontsize=9,
    color='red',
    ha='center'
)

plt.tight_layout()
plt.show()
# Histogram + KDE
import seaborn as sns

plt.figure(figsize=(10, 5),dpi=300)

# Histogram with KDE
sns.histplot(df['Price'], kde=True, bins=20, color='brown', label='Price Distribution')

# Titles and labels
plt.title('Price Distribution', fontsize=16)
plt.xlabel('Price ($)', fontsize=16)
plt.ylabel('Count', fontsize=16)

# Ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Legend on top right
plt.legend(loc='upper right', fontsize=16)

plt.tight_layout()
plt.show()
# Monthly average
# Calculate monthly average if not already done
monthly_avg = df['Price'].resample('M').mean()

plt.figure(figsize=(10, 5))
plt.bar(monthly_avg.index, monthly_avg.values, color='brown', width=20)  # width=20 for better bar visibility
plt.title('Monthly Average Oil Prices')
plt.xlabel('Month')
plt.ylabel('Price ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Violin plot function
def visualize_yearly_violin(data):
    """
    Violin plot of oil prices per year.

    Parameters:
        data (pd.DataFrame): DataFrame with 'Price' and datetime index.
    """
    df_temp = data.copy()
    df_temp['Year'] = df_temp.index.year

    # Yearly violin plot
    plt.figure(figsize=(10, 5))
    sns.violinplot(x='Year', y='Price', data=df_temp, palette='hsv')
    plt.title("Yearly Oil Price Distribution", fontsize=16)
    plt.xlabel("Year")
    plt.ylabel("Price ($)")
    plt.grid(True)
    plt.show()

# Call the function
visualize_yearly_violin(df)
```

```python
# ADF test on original series
result = adfuller(df['Price'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
# First difference
df_diff = df['Price'].diff().dropna()
# ADF on differenced data
result_diff = adfuller(df_diff)
print('ADF after differencing:', result_diff[0])
print('p-value:', result_diff[1])
# ACF and PACF
plot_acf(df_diff)
plot_pacf(df_diff)
plt.show()
```

```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore")

def Arima(Series, p_range, d_range, q_range):

    Best_aic = np.inf
    Best_order = None

    for p in p_range:
        for d in d_range:
            for q in q_range:
                try:
                    model = ARIMA(Series, order=(p, d, q))
                    result = model.fit()
                    if result.aic < Best_aic:
                        Best_aic = result.aic
                        Best_order = (p, d, q)
                except:
                    continue
    return Best_order, Best_aic

# Define parameter ranges
p_values = range(0, 9)
d_values = range(0, 3)
q_values = range(0, 9)

# Run grid search
Best_order, Best_aic = Arima(df['Price'], p_values, d_values, q_values)

print("Arima Order Found:", Best_order)
print("Corresponding AIC:", round(Best_aic, 2))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
from math import sqrt
import warnings

warnings.filterwarnings("ignore")

# -------- Function: SARIMA Model Forecasting --------
def sarima_model_forecasting(model_fit, steps_ahead=720, last_date=None):
    forecast = model_fit.get_forecast(steps=steps_ahead)
    forecast_mean = forecast.predicted_mean
    conf_int = forecast.conf_int()

    forecast_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=steps_ahead, freq='D')
    forecast_mean.index = forecast_index
    conf_int.index = forecast_index

    return forecast_mean, conf_int, forecast_index

# Download Actual Prices 
def download_actual_prices(start_date, end_date):
    DATA = yf.download("BZ=F", start=start_date, end=end_date)
    if 'Adj Close' in DATA.columns:
        prices = DATA[['Adj Close']].rename(columns={'Adj Close': 'Price'})
    elif 'Close' in DATA.columns:
        prices = DATA[['Close']].rename(columns={'Close': 'Price'})
    else:
        raise ValueError("No valid price column found in Yahoo Finance data.")
    prices.index = pd.to_datetime(prices.index)
    prices = prices.asfreq('D').interpolate()
    return prices

#  RMSE Evaluation
def evaluate_forecast(actual_series, forecast_series):
    df_eval = pd.DataFrame({
        'actual': actual_series.squeeze(),
        'forecast': forecast_series.squeeze()
    }).dropna()
    return sqrt(mean_squared_error(df_eval['actual'], df_eval['forecast']))

# Load the dataset (assumes df is already prepared)
seasonal_order = (1, 1, 1, 7)  # Example: weekly seasonality, tune as needed
sarima_order = (2, 1, 2)       # Example: tune as needed

sarima_model = SARIMAX(df['Price'], order=sarima_order, seasonal_order=seasonal_order).fit()

# Forecasting the future prices
forecast_mean, conf_int, forecast_index = sarima_model_forecasting(
    sarima_model, steps_ahead=720, last_date=df.index[-1]
)

# Download real prices for evaluation
start_date = forecast_index[0].strftime('%Y-%m-%d')
end_date = pd.to_datetime("today").strftime('%Y-%m-%d')
actual_prices_df = download_actual_prices(start_date, end_date)

# Align and calculate RMSE
common_dates = actual_prices_df.index.intersection(forecast_mean.index)
actual_series = actual_prices_df['Price'].reindex(common_dates).squeeze()
predicted_series = forecast_mean.reindex(common_dates).squeeze()
rmse_score = evaluate_forecast(actual_series, predicted_series)

# Plot Forecast vs Actual
plt.figure(figsize=(14, 6))
plt.plot(df['Price'], label="Historical Price", color='brown')
plt.plot(forecast_index, forecast_mean, label="SARIMA Forecast", color='green')
plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='lightgreen', alpha=0.3, label="95% Confidence Interval")
plt.plot(actual_prices_df['Price'], '--', color='black', label="Actual Price (Yahoo)")
plt.title(f"Oil Price Forecast vs Actual (SARIMA order={sarima_order}, seasonal={seasonal_order})")
plt.xlabel("Date")
plt.ylabel("Oil Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Root Mean Squared Error (RMSE): {rmse_score:.2f} USD")
```
